```python
import os
import sys

import pandas as pd

# Path del dataset
path = "../../data/dataset/applets/applets_real_new.jsonl"

# Lista degli indici del tuo study set
novice_indices = [90, 84, 76, 66, 88, 92]
expert_indices = [218, 207, 238, 116, 231, 204]

study_indices = novice_indices + expert_indices

# Carica dataset
df = pd.read_json(path, lines=True)

# Filtra gli applet
df_study = df[df["row_index"].isin(study_indices)].copy()

# Ordina per comodità
df_study = df_study.sort_values("row_index")

# Mostra risultato
print(df_study[["row_index", "name"]])

# Se vuoi esportare:
df_study.to_json("study_set_applets.json", orient="records", indent=2)
```
